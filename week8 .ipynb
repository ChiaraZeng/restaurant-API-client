```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("combined_restaurants_cleaned.csv")

# Select features for clustering and remove missing values
X = df[['rating', 'price_level']].dropna()

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

# Create a new DataFrame with cluster labels
df_clustered = df.loc[X.index].copy()
df_clustered['cluster'] = clusters

# Plot results
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_clustered, x='rating', y='price_level', hue='cluster', palette='Set1', s=100)
plt.title('K-Means Clustering of Restaurants by Rating and Price Level')
plt.xlabel('Rating')
plt.ylabel('Price Level')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()

```
