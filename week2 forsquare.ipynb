```python
import requests
import json

# Function to query Foursquare Places API
def fetch_foursquare_restaurants(lat, lon, api_key, radius=5000, limit=10):
    url = "https://api.foursquare.com/v3/places/search"
    headers = {
        "Accept": "application/json",
        "Authorization": api_key
    }
    params = {
        "query": "restaurant",
        "ll": f"{lat},{lon}",   # Format: "latitude,longitude"
        "radius": radius,       # In meters
        "limit": limit
    }

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    results = data.get('results', [])
    for r in results:
        name = r.get('name')
        address = r.get('location', {}).get('formatted_address', 'No address')
        print(f"{name} – {address}")

    return results

# === Replace with your actual Foursquare API key ===
api_key = "fsq3RZiQAOeHzAEslLwurjhnkL23N8jH+CwkZILirp836wg="

# Coordinates for each city (latitude, longitude)
cities = {
    "New York": (40.7128, -74.0060),
    "San Francisco": (37.7749, -122.4194),
    "Chicago": (41.8781, -87.6298),
    "Boston": (42.3601, -71.0589),
    "Los Angeles": (34.0522, -118.2437)
}

radius = 5000  # Search radius in meters (e.g., 5000 = 5km)

# Fetch, print, and save results for each city
for city, (lat, lon) in cities.items():
    print(f"\n--- Restaurants in {city} ---")
    restaurant_data = fetch_foursquare_restaurants(lat, lon, api_key, radius)

    # Save to JSON file
    with open(f"{city.replace(' ', '_').lower()}_restaurants_foursquare.json", 'w') as f:
        json.dump(restaurant_data, f, indent=2)
```
