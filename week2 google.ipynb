```python
import requests
import json

def fetch_google_restaurants(city, api_key, limit=10):
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
    params = {
        'query': f'restaurants in {city}',
        'key': api_key
    }
    response = requests.get(url, params=params)
    data = response.json()
    results = data.get('results', [])[:limit]

    for r in results:
        print(f"{r['name']} – {r['formatted_address']}")

    return results

# Replace with your actual API key
api_key = "AIzaSyD6ISYFH3oUWXUvGpbxpOaLlhW3u-R4q7w"

cities = ['New York', 'San Francisco', 'Chicago', 'Boston', 'Los Angeles']

# Fetch and save for each city
for city in cities:
    print(f"\nFetching restaurants in {city}...")
    restaurant_data = fetch_google_restaurants(city, api_key)

    with open(f"{city.replace(' ', '_').lower()}_restaurants_google.json", 'w') as f:
        json.dump(restaurant_data, f, indent=2)
```
